```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.cluster import DBSCAN
```

```python
X, Y = make_blobs(n_samples=250, n_features=2, random_state=40)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y, cmap='rainbow', edgecolors='black', s=100)
```

```python
clf = DBSCAN(eps=1, min_samples=7)
```

```python
clf.fit(X,Y)
```

```python
labels = clf.labels_.tolist()
```

```python
for i in range(len(labels)):
    if labels[i] == 0:
        plt.scatter(X[i,0], X[i,1], c='red', s=100, label='C1')
    elif labels[i] == 1:
        plt.scatter(X[i,0], X[i,1], c='blue', s=100, label='C2')
    elif labels[i] == 2:
        plt.scatter(X[i,0], X[i,1], c='green', s=100, label='C3')
    elif labels[i] == -1:
        plt.scatter(X[i,0], X[i,1], c='black', s=100, label='Noise')
```
