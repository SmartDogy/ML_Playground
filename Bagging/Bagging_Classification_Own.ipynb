```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.tree import DecisionTreeClassifier
from sklearn.metrics import accuracy_score
from scipy import stats
```

```python
dataset_raw = load_iris()
```

```python
dataset = pd.DataFrame(data = dataset_raw['data'], columns = dataset_raw['feature_names'])
```

```python
dataset.head()
```

```python
dataset['target'] = dataset_raw['target']
```

```python
dataset.head()
```

```python
dataset = dataset.drop('sepal length (cm)', axis = 1)
```

```python
dataset = dataset.drop('petal length (cm)', axis = 1)
```

```python
dataset.head()
```

```python
dataset = dataset[dataset['target'] != 0]
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
dataset = dataset.sample(100)
```

```python
dataset = dataset.reset_index(drop = True)
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, [0,1]].values
Y = dataset.iloc[:, -1].values
```

```python
x_train_full = X[:70]
y_train_full = Y[:70]

x_test_full = X[70:]
y_test_full = Y[70:]
```

```python
x_train_full.shape
```

```python
accuracy = []
prediction = []

for i in range(5):

  index = np.random.choice(70, 50, replace=True)
  x_train_small = x_train_full[index]
  y_train_small = y_train_full[index]

  clf = DecisionTreeClassifier()
  clf.fit(x_train_small, y_train_small)

  accuracy.append(accuracy_score(y_test_full, clf.predict(x_test_full)))

  output = 1.2, 1.1
  prediction.append(clf.predict(([output])))
```

```python
print(accuracy)
print(prediction)

#prediction = np.concatenate(prediction)
final_pred = stats.mode(prediction).mode[0]

print(final_pred)
```
