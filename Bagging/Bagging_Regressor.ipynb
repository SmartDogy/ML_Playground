```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import BaggingRegressor
from sklearn.metrics import accuracy_score
from sklearn.model_selection import train_test_split
```

```python
dataset = pd.read_csv('/content/CGPA_LPA_500.csv')
```

```python
dataset.head()
```

```python
plt.scatter(dataset['CGPA'], dataset['LPA'], s=100, edgecolors='black')
plt.xlabel('CGPA')
plt.ylabel('LPA')
plt.grid()
```

```python
X = dataset[['CGPA']].values
Y = dataset[['LPA']].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.3, random_state=42)
```

```python
lr = LinearRegression()
lr.fit(x_train, y_train)
```

```python
lr.predict([[4.5]])
```

```python
dt = DecisionTreeRegressor()
dt.fit(x_train, y_train)
```

```python
dt.predict([[4.5]])
```

```python
x_train.shape
```

```python
clf = BaggingRegressor(estimator=DecisionTreeRegressor(), n_estimators=100, max_samples=250, bootstrap=True, random_state=42)
```

```python
clf.fit(x_train, y_train)
```

```python
clf.predict([[4.5]])
```
