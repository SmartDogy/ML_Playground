```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.datasets import make_blobs
from sklearn.svm import SVC
```

```python
X, Y = make_blobs(n_samples=40, centers=2, random_state=2)
```

```python
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='winter', s=100, edgecolors='black')
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid()
```

```python
clf = SVC(kernel='linear', C=2)
clf.fit(X,Y)
```

Screenshot 2025-08-15 at 6.04.30â€¯PM.png

```python
w = clf.coef_[0]
b = clf.intercept_[0]
x_vals = np.linspace(X[:, 0].min(), X[:, 0].max(), 100)
y_vals = -(w[0] / w[1]) * x_vals - b / w[1]
plt.scatter(X[:, 0], X[:, 1], c=Y, cmap='winter', s=100, edgecolors='black')
plt.plot(x_vals, y_vals, label='Decision Boundary')
plt.legend()
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.grid()
```
