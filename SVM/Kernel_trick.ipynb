```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import r2_score
from sklearn.datasets import make_circles
from mlxtend.plotting import plot_decision_regions
```

```python
X, Y = make_circles(n_samples=200, factor=0.1, noise=0.1)
```

```python
plt.scatter(X[:,0], X[:,1], c=Y, cmap='winter',edgecolors='black')
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
clf = SVC(kernel='rbf')
clf.fit(x_train, y_train)
```

```python
print(r2_score(y_test, clf.predict(x_test)))
```

```python
plot_decision_regions(X,Y,clf)
plt.grid()
```

To plot the datapoints in 2D

we use the formula e^−x^2

```python
newX = np.exp(-(X ** 2))
```

```python
plt.scatter(newX[:, 0], newX[:, 1], c=Y, cmap='winter', edgecolors='black')
```
