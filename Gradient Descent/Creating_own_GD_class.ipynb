```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
X, Y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=30)
```

```python
plt.scatter(X,Y)
```

```python
clf = LinearRegression()
clf.fit(X,Y)
plt.scatter(X,Y)
plt.plot(X,clf.predict(X), color='black')
```

```python
print(clf.coef_)
print(clf.intercept_)
```

```python
class GDregressor:

  def __init__(self, lr, epochs):
    self.m = 59.3711242
    self.b = 0
    self.lr = lr
    self.epochs = epochs

  def fit(self, X, Y):
      for i in range(self.epochs):
        slope = -2 * np.sum(Y - self.m*X.ravel() - self.b)
        self.b = self.b - (self.lr * slope)

      print(self.b)
```

```python
GD = GDregressor(0.001, 100)
```

```python
GD.fit(X, Y)
```
