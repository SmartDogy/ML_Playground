```python
from sklearn.datasets import make_regression
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
```

```python
X, Y = make_regression(n_samples=100, n_features=1, n_informative=1, n_targets=1, noise=30)
```

```python
plt.scatter(X,Y)
```

```python
clf = LinearRegression()
clf.fit(X,Y)
pred_val = clf.predict(X)
plt.scatter(X, Y)
plt.plot(X, pred_val, color='black', label='Linear Regression')
plt.legend()
```

```python
print(clf.coef_)
print(clf.intercept_)
```

```python
class GradientRegressor:

  def __init__(self, lr, epochs):
    self.m = 1
    self.b = 0
    self.lr = lr
    self.epochs = epochs

  def fit(self, X, Y):
    for i in range(self.epochs):
      slope_b = -2 * np.sum(Y - self.m * X.ravel() - self.b)
      slope_m = -2 * np.sum((Y - self.m * X.ravel() - self.b) * X.ravel())
      self.m = self.m - (self.lr * slope_m)
      self.b = self.b - (self.lr * slope_b)
    print(self.m, self.b)
```

```python
GD = GradientRegressor(0.001, 100)
```

```python
GD.fit(X,Y)
```
