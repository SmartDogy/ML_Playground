```python
class Stochastic_Gradient_Descent_Regressor:

  def __init__(self, lr=0.01, epochs=10):
    self.lr = lr
    self.epochs = epochs
    self.coef = None
    self.intercept = None

  def fit(self, x_train, y_train):
    self.intercept = 0
    self.coef = np.ones(x_train.shape[1])

    for i in range(self.epochs):

      for j in range(x_train.shape[0]):

        index = np.random.randint(0, x_train.shape[0])

        y_hat = np.dot(x_train[index], self.coef) + self.intercept

        error = y_train[index] - y_hat

        slope_intercept = -2 * (error)
        self.intercept = self.intercept - (self.lr * slope_intercept)

        slope_coef = -2 * (np.dot(error, x_train[index]))
        self.coef = self.coef - (self.lr * slope_coef)

    print(self.intercept, self.coef)

  def predict(self, x_test):
    return np.dot(x_test, self.coef) + self.intercept
```

```python
SGDR = Stochastic_Gradient_Descent_Regressor(0.01, 50)
SGDR.fit(x_train, y_train)
```

```python
y_pred = SGDR.predict(x_test)
```

```python
r2_score(y_test, y_pred)
```
