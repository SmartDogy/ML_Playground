```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.preprocessing import LabelEncoder
from sklearn.datasets import load_iris
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import r2_score
from sklearn.ensemble import VotingClassifier
```

```python
dataset_ = load_iris()
```

```python
dataset = pd.DataFrame(data=dataset_['data'], columns=dataset_['feature_names'])
```

```python
dataset.head()
```

```python
LE = LabelEncoder()
LE.fit_transform(dataset_['target'])
```

```python
dataset['Target'] = dataset_['target']
```

```python
dataset.head()
```

```python
dataset = dataset.iloc[:, [0,1,4]]
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
X = dataset.iloc[:,[0,1]].values
Y = dataset.iloc[:,-1].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
clf1 = LogisticRegression()
clf2 = KNeighborsClassifier()
clf3 = RandomForestClassifier()
```

```python
estimator = [('lr', clf1), ('knn', clf2), ('rf', clf3)]
```

```python
vc = VotingClassifier(estimators=estimator, voting='soft')
```

```python
vc.fit(x_train, y_train)
```

```python
r2_score(y_test, vc.predict(x_test))
```

```python
accuracy = []
clf1.fit(x_train, y_train)
accuracy.append(r2_score(y_test,clf1.predict(x_test)))
clf2.fit(x_train, y_train)
accuracy.append(r2_score(y_test,clf2.predict(x_test)))
clf3.fit(x_train, y_train)
accuracy.append(r2_score(y_test,clf3.predict(x_test)))
np.mean(accuracy)
```

```python
acc = []

for i in range(1, 11):
  knn = KNeighborsClassifier(n_neighbors=i)
  knn.fit(x_train, y_train)
  acc.append(r2_score(y_test, knn.predict(x_test)))

acc
```

```python
knn1 = KNeighborsClassifier(n_neighbors=1)
knn2 = KNeighborsClassifier(n_neighbors=2)
knn3 = KNeighborsClassifier(n_neighbors=3)
knn4 = KNeighborsClassifier(n_neighbors=4)
knn5 = KNeighborsClassifier(n_neighbors=5)

est = [('knn1', knn1), ('knn2', knn2), ('knn3', knn3), ('knn4', knn4), ('knn5', knn5)]
clf = VotingClassifier(estimators=est, voting='soft')
clf.fit(x_train, y_train)
```

```python
print(r2_score(y_test, clf.predict(x_test)))
```
