```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.cluster import AgglomerativeClustering
from sklearn.model_selection import train_test_split
from scipy.cluster import hierarchy
```

```python
dataset = pd.read_csv('/content/blob_clusters.csv')
```

```python
dataset.head()
```

```python
plt.scatter(dataset['CGPA'], dataset['IQ'], c=dataset['Placement'], s=100, edgecolors='black', cmap='bwr')
```

```python
X = dataset.iloc[:, 0:2].values
Y = dataset.iloc[:, 2].values
```

```python
x_train, y_train, x_test, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
plt.figure(figsize=(10, 7))
dend = hierarchy.dendrogram(hierarchy.linkage(dataset, method='ward'))
```

```python
clf = AgglomerativeClustering(n_clusters=2, linkage='ward')
labels_ = clf.fit_predict(X, Y)
```

```python
plt.scatter(X[:,0], X[:,1], c=clf.labels_, cmap='rainbow')
```
