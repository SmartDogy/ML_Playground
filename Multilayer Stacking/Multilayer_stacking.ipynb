```python
import pandas as pd
import numpy as np

from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier

from sklearn.ensemble import GradientBoostingClassifier
from sklearn.ensemble import AdaBoostClassifier
from sklearn.tree import DecisionTreeClassifier

from sklearn.ensemble import RandomForestClassifier

from sklearn.model_selection import train_test_split

from sklearn.metrics import accuracy_score
```

```python
data = pd.read_csv('/content/1000cgpa_iq.csv')
```

```python
data.shape
```

```python
data.head()
```

```python
lr = LogisticRegression()
svc = SVC()
knn = KNeighborsClassifier(n_neighbors=5)
```

```python
gb = GradientBoostingClassifier()
ab = AdaBoostClassifier()
dt = DecisionTreeClassifier()
```

```python
rf = RandomForestClassifier()
```

```python
X = data.iloc[:, 0:2].values
Y = data.iloc[:, 2].values
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2)
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
x_test.shape
```

```python
y_test.shape
```

```python
# Fold 1
x_train_fold1 = x_train[0:320, :]
y_train_fold1 = y_train[0:320]

# Fold 2
x_train_fold2 = x_train[320:640, :]
y_train_fold2 = y_train[320:640]

# Fold 3
x_train_fold3 = x_train[640:960, :]
y_train_fold3 = y_train[640:960]

# Fold 4
x_train_fold4 = x_train[960:1280, :]
y_train_fold4 = y_train[960:1280]

# Fold 5
x_train_fold5 = x_train[1280:1600, :]
y_train_fold5 = y_train[1280:1600]
```

# Layer 1 training

```python
lr_pred = np.array(lr_pred)
```

```python
lr_pred = np.concatenate(lr_pred)
```

```python
lr_pred.shape
```

```python
#KNN

knn_pred = []

knn.fit(x_train_fold1, y_train_fold1)
knn.fit(x_train_fold2, y_train_fold2)
knn.fit(x_train_fold3, y_train_fold3)
knn.fit(x_train_fold4, y_train_fold4)

knn_pred.append(knn.predict(x_train_fold5))

knn.fit(x_train_fold1, y_train_fold1)
knn.fit(x_train_fold2, y_train_fold2)
knn.fit(x_train_fold3, y_train_fold3)
knn.fit(x_train_fold5, y_train_fold5)

knn_pred.append(knn.predict(x_train_fold4))

knn.fit(x_train_fold1, y_train_fold1)
knn.fit(x_train_fold2, y_train_fold2)
knn.fit(x_train_fold4, y_train_fold4)
knn.fit(x_train_fold5, y_train_fold5)

knn_pred.append(knn.predict(x_train_fold3))

knn.fit(x_train_fold1, y_train_fold1)
knn.fit(x_train_fold3, y_train_fold3)
knn.fit(x_train_fold4, y_train_fold4)
knn.fit(x_train_fold5, y_train_fold5)

knn_pred.append(knn.predict(x_train_fold2))

knn.fit(x_train_fold2, y_train_fold2)
knn.fit(x_train_fold3, y_train_fold3)
knn.fit(x_train_fold4, y_train_fold4)
knn.fit(x_train_fold5, y_train_fold5)

knn_pred.append(knn.predict(x_train_fold1))
```

```python
layer_2_dataset = pd.DataFrame()
layer_2_dataset['LR_PRED'] = lr_pred
layer_2_dataset['SVC_PRED'] = svc_pred
layer_2_dataset['KNN_PRED'] = knn_pred
layer_2_dataset['Placement'] = y_train

layer_2_dataset.head()
```
