```python
dataset_df = pd.DataFrame(data = dataset)
```

```python
le = LabelEncoder()
for col in dataset_df.columns:
  dataset_df[col] = le.fit_transform(dataset_df[col])
```

```python
X = dataset_df.iloc[:, [0,1,2,3]].values
Y = dataset_df.iloc[:, 4].values
```

```python
x_train, x_test, y_train, y_ttest = train_test_split(X, Y, test_size=0.3)
```

```python
clf = GaussianNB()
```

```python
clf.fit(x_train, y_train)
```

```python
y_pred = clf.predict(x_test)
```

```python
accuracy_score(y_ttest, y_pred)
```
