```python
import pandas as pd
import numpy as np
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import BaggingClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score
from sklearn.datasets import make_classification
from scipy import stats
from mlxtend.plotting import plot_decision_regions
```

```python
X, Y = make_classification(n_samples=500, n_features=5, n_informative=5, n_redundant=0)
```

```python
Y.shape
```

```python
dataset = pd.DataFrame(data=X)
```

```python
dataset.head()
```

```python
dataset['target'] = Y
```

```python
dataset.head()
```

```python
x_train, x_test, y_train, y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
dt = DecisionTreeClassifier(max_depth=10)
dt.fit(x_train, y_train)
```

```python
accuracy_score(y_test, dt.predict(x_test))
```

```python
x_train.shape
```

```python
clf = BaggingClassifier(estimator=DecisionTreeClassifier(), n_estimators=100, max_samples=250, bootstrap=True)
```

```python
clf.fit(x_train, y_train)
```

```python
accuracy_score(y_test, clf.predict(x_test))
```

```python
accuracy = []
prediction = []

for i in range(9):

  index =np.random.choice(400, 250, replace = True)

  x_train_small = x_train[index]
  y_train_small = y_train[index]

  random_forest = DecisionTreeClassifier()
  random_forest.fit(x_train_small, y_train_small)

  accuracy.append(accuracy_score(y_test, random_forest.predict(x_test)))
  val = -0.143170, 0.891891, 0.913304,	0.685793,	0.705082
  prediction.append(random_forest.predict([val]))
```

```python
print(accuracy)
print(prediction)

print(prediction)

result = stats.mode(prediction).mode[0]

print(result)
```
