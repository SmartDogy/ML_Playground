```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.datasets import make_regression
```

```python
X, Y = make_regression(n_samples=4, n_features=1,n_informative=1, n_targets=1, noise=80, random_state=13)
```

```python
plt.scatter(X, Y)
```

```python
#Applying OLS(Ordinar Least Square)
clf = LinearRegression()
clf.fit(X, Y)
```

```python
clf.coef_
```

```python
clf.intercept_
```

```python
plt.scatter(X, Y)
plt.plot(X, clf.predict(X), color='red')
```

Now we will use Gradient Descent and apply our logic and get close to the real intercept(b) which is 26.15963284313262

Now we will start with our b with 0 and we are given that m which is constant is 78.35

Formula:

y_hat = mx + b

we given that m = 78.35 and we asuming b = 0

∴ y_hat = 78.35 * X + b

```python
b = 0
m = 78.35
y_pred = (m * X) + 0
```

```python
plt.scatter(X,Y)
plt.plot(X, clf.predict(X), color='red',label='OLS')
plt.plot(X, y_pred, color='green',label='b=0')
plt.legend()
```

Now I have to find slope so that I can get closse to red line whichs is OLS line

Formula to calculate slope:

We get this formula after defrentiating Σ(yi - y_hat)^2

(y_hat is nothing but mx + b)

So hence slope would be

-2Σ(yi - mxi - b)


```python
slope = -2 * np.sum((Y-(m*X.ravel())-b)) #.ravel() function needed when there is 2D array
```

So at last we know that bnew = bold - lr(slope)

So we have slope also and lets take lr = 0.1

and the whole thing lr(slope) is called step size

```python
lr = 0.1
```

```python
b = b - (lr * slope)
```

```python
b
```

```python
y_pred2 = (m * X) + b
```

```python
plt.scatter(X, Y)
plt.plot(X, clf.predict(X), color='red', label='OLS')
plt.plot(X, y_pred2, color='black', label='b = 20.927')
plt.legend()
```

```python
slope = -2 * (np.sum(Y-(m*X.ravel())-b))
slope
```

```python
step_size = lr*slope
```

```python
step_size
```

```python
b = b - step_size
```

```python
y_pred3 = (m * X) + b
```

```python
plt.scatter(X, Y)
plt.plot(X, clf.predict(X), color='red', label='OLS')
plt.plot(X, y_pred2, color='black', label='b = 20.927')
plt.plot(X, y_pred3, color='green', label='b = 20.927')
plt.legend()
```

```python
m = 78.35
b = -100
lr = 0.1
epochs = 10

for i in range(epochs):
  slope = -2 * np.sum(Y - m*X.ravel() - b)
  b = b - (lr * slope)
  y_pred = (m * X) + b

  plt.plot(X, y_pred)
plt.scatter(X,Y)
```
