```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import  LinearRegression
```

```python
dataset = pd.read_csv('/content/salary_pacakge.csv')
```

```python
dataset.head()
```

```python
X = dataset.iloc[:, 1].values.reshape(-1, 1)
Y = dataset.iloc[:, 2].values.reshape(-1, 1)
```

```python
plt.scatter(X, Y)
plt.xlabel('CGPA')
plt.ylabel('Package')
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 2 )
```

```python
lr = LinearRegression()
```

```python
lr.fit(X_train, Y_train)
```

```python
lr.predict(X_test[[0]])
```

```python
User_data = float(input('Enter your years of experienence: '))
if(User_data >= 50):
  print('Your years of experience cannot exceed 50')
elif(User_data <= -1):
  print("Your years of experience cannot exceed 0")
else:
  User_data = np.array([[User_data]])
  prediction = lr.predict(User_data)
  print(f'Your salary is {round(prediction[0][0], 1)} LPA')
```
